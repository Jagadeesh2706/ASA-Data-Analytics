```python
# prompt: Summarize the target variable (Churn) distribution.generate the code

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('customer_data.csv')

# Assuming 'df' is your DataFrame (already loaded in the provided code)

# Method 1: Using seaborn countplot
plt.figure(figsize=(8, 5))
sns.countplot(x=df['Churn'], palette='viridis')
plt.xlabel('Churn')
plt.ylabel('Frequency')
plt.title('Churn Distribution')
plt.xticks(rotation=0)
plt.show()

# Method 2: Using pandas value_counts and plotting
df['Churn'].value_counts().plot(kind='bar', color=['red', 'green'])
plt.title('Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.show()

# Method 3:  Print the value counts directly
print(df['Churn'].value_counts())
```

```python
# prompt: 1.	Handle missing values:
# o	Impute numerical columns with median or mean.

# Identify numerical columns with missing values
numerical_cols_with_missing = df.select_dtypes(include=['number']).columns[df.select_dtypes(include=['number']).isnull().any()]

# Impute missing values in numerical columns using the median
for col in numerical_cols_with_missing:
    median_val = df[col].median()
    df[col].fillna(median_val, inplace=True)

# Verify if missing values are handled
print(df.isnull().sum())
```

```python
# prompt: o	Impute categorical columns with the mode.

# Identify categorical columns with missing values
categorical_cols_with_missing = df.select_dtypes(include=['object']).columns[df.select_dtypes(include=['object']).isnull().any()]

# Impute missing values in categorical columns using the mode
for col in categorical_cols_with_missing:
    mode_val = df[col].mode()[0]  # Get the first mode in case of multiple modes
    df[col].fillna(mode_val, inplace=True)

# Verify if missing values are handled
print(df.isnull().sum())
```

```python
# prompt: 2.	Remove duplicate rows, if any.

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Verify if duplicates are removed
print(f"Number of duplicate rows after removal: {df.duplicated().sum()}")
```

```python
# prompt: 3.	Convert all categorical features into numeric using one-hot encoding.

import pandas as pd
# Get a list of categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Perform one-hot encoding
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Display the updated DataFrame
print(df.head())
```

```python
# prompt: 1.	Create new features from existing columns:
# o	Tenure_Group: Group tenure into categories (e.g., 0-12 months, 13-24 months, etc.).

import pandas as pd
# Define the bins for tenure groups
bins = [0, 12, 24, 36, 48, 60, 72]  # Example bins (adjust as needed)
labels = ['0-12 months', '13-24 months', '25-36 months', '37-48 months', '49-60 months', '61-72 months']

# Create the 'Tenure_Group' column using pd.cut
df['Tenure_Group'] = pd.cut(df['tenure'], bins=bins, labels=labels, right=False)

# Display the updated DataFrame to see the new feature
print(df.head())
```

```python
# prompt: 1.	Create new features from existing columns:
# o	Average_Monthly_Charges: Calculate Total Charges / Tenure.

# Calculate 'Average_Monthly_Charges'
df['Average_Monthly_Charges'] = df['TotalCharges'] / df['tenure']

# Display the updated DataFrame to see the new feature
print(df.head())
```

```python
# prompt: 2.	Create interaction features between important numerical columns (e.g., Monthly Charges * Tenure).

# Create interaction features
df['MonthlyCharges_Times_Tenure'] = df['MonthlyCharges'] * df['tenure']
df['TotalCharges_Divided_Tenure'] = df['TotalCharges'] / df['tenure']

# Display the updated DataFrame to see the new features
print(df.head())
```

```python
# prompt: 3.	Perform scaling or normalization for numerical features

from sklearn.preprocessing import MinMaxScaler

# Assuming 'df' is your DataFrame and 'numerical_features' is a list of numerical column names

# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit the scaler to your numerical features and transform them
df[numerical_features] = scaler.fit_transform(df[numerical_features])

# Print the scaled DataFrame
print(df.head())
```
